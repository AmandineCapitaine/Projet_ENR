### 1. Utils

```python
def Profil_annuel(profil, vre, verbose=False) :
    if vre=="offshore":
        n=0
    elif vre=="onshore":
        n=8759
    elif vre=="pv":
        n=17519
    profil_annee = np.zeros(8760)
    for i in range(8760):
        profil_annee[i-1] += [profil["prod"][i+n]]
    plt.plot(profil_annee)
    plt.title("Profil de production de la technologie : " + vre)
    plt.xlabel("Heures de l'année")
```

### 2. Profils de l'éolien offshore

```python
Profil_annuel(vre2017, "offshore")
```

### 3. Profils de l'éolien onshore

```python
Profil_annuel(vre2017, "onshore")
```

### 4. Profils du solaire

```python
Profil_annuel(vre2017, "pv")
```
